# **2. Level of Preparedness of Alumni that the Accounting Curriculum Provides**

2.1 Theoretical Knowledge

```python
# Assuming 'df' is your DataFrame from the previous code

def calculate_nar(column):
    strongly_agree = column[column == 4].count()
    agree = column[column == 3].count()
    disagree = column[column == 2].count()
    strongly_disagree = column[column == 1].count()
    total_responses = column.count()
    nar = (((strongly_agree + agree) - (strongly_disagree + disagree)) / total_responses) * 100
    return nar

# Calculate median and IQR for THK1, THK2, THK3
columns_to_analyze = ['THK1', 'THK2', 'THK3']
results = []

for col in columns_to_analyze:
    median = df[col].median()
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    nar = calculate_nar(df[col])
    results.append([col, median, iqr, nar])

# Calculate overall median and IQR
overall_median = df[columns_to_analyze].median().median()
overall_q1 = df[columns_to_analyze].quantile(0.25).min()
overall_q3 = df[columns_to_analyze].quantile(0.75).max()
overall_iqr = overall_q3 - overall_q1
overall_nar = calculate_nar(pd.concat([df[col] for col in columns_to_analyze])) #overall NAR

results.append(['Overall', overall_median, overall_iqr, overall_nar])


# Display results in a table
print("{:<10} {:<10} {:<10} {:<10}".format("Column", "Median", "IQR", "NAR"))
for result in results:
    print("{:<10} {:<10.2f} {:<10.2f} {:<10.2f}".format(result[0], result[1], result[2], result[3]))
```
