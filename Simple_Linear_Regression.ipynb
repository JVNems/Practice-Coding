# **I. Importing Relevant Modules for Regression**

```python
# Import necessary libraries
import pandas as pd  # Used for data manipulation and analysis
import numpy as np  # Used for numerical operations
import matplotlib.pyplot as plt  # Used for data visualization
import seaborn as sns  # Used for enhanced data visualization
from sklearn.model_selection import train_test_split  # Used for splitting data into training and testing sets
from sklearn.linear_model import LinearRegression  # Used for linear regression modeling
from sklearn.metrics import mean_squared_error, r2_score  # Used for evaluating regression model performance
```

# **II. Importing CSV Files**

```python
# prompt: i want to import files csv /content/IceCreamData.csv


# Load the CSV file into a pandas DataFrame.
df = pd.read_csv('/content/IceCreamData.csv')
```

```python
# prompt: calculate the average and maximun of temperature and revenue

# Calculate the average temperature
average_temp = df['Temperature'].mean()
print(f"Average Temperature: {average_temp}")

# Calculate the maximum temperature
max_temp = df['Temperature'].max()
print(f"Maximum Temperature: {max_temp}")

# Calculate the average revenue
average_revenue = df['Revenue'].mean()
print(f"Average Revenue: {average_revenue}")

# Calculate the maximum revenue
max_revenue = df['Revenue'].max()
print(f"Maximum Revenue: {max_revenue}")
```

# **IV.Visualize Data**

```python
# prompt: create a scatter plot, and add line correspondin the plot

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature', y='Revenue', data=df, color='blue', label='Data Points')

# Fit a linear regression model
X = df[['Temperature']]
y = df['Revenue']
model = LinearRegression()
model.fit(X, y)

# Predict revenue based on temperature
y_pred = model.predict(X)

# Add the regression line to the plot
plt.plot(X, y_pred, color='red', label='Regression Line')

# Customize the plot
plt.title('Temperature vs. Revenue')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

# **VI. Simple Linear Regression**

```python
# prompt: perform simple linear regression using sklearn, and display necessaty values

# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Assuming X and Y are already defined as in the provided code
# X = df[['Temperature']]
# Y = df['Revenue']
# X = np.array(X)
# Y = np.array(Y)
# X = X.reshape(-1, 1)
# Y = Y.reshape(-1, 1)

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Make predictions
Y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

# Print the model coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
