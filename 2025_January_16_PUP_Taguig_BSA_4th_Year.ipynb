```python
# Create a simple bar chart for the 'Age' column
plt.figure(figsize=(8, 5))
ax = df['Age'].value_counts().sort_index().plot(kind='bar', color='blue', edgecolor='black')
plt.title('Age Distribution', fontsize=14)
plt.xlabel('')  # Set the x-axis label to an empty string
plt.ylabel('')  # Set the y-axis label to an empty string

# Customize x-axis labels
labels = {1: '18-20 years old', 2: '21-23 years old', 3: '24-26 years old'}
plt.xticks(ticks=range(len(labels)), labels=[labels.get(x, str(x)) for x in range(1, 4)], rotation=0)


plt.grid(axis='y', linestyle='', alpha=0.7)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add frequency and percentage above each bar
total = len(df['Age'])
for p in ax.patches:
    height = p.get_height()
    percentage = '{:.2f}%'.format(100 * height / total)  # Format percentage to 2 decimal places
    ax.text(p.get_x() + p.get_width() / 2, height + 1,
            f'{int(height)} ({percentage})',  # Change the text format
            ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# prompt: calculate the median and interquartile range of A1, A2,A3,A4,A5 column. sHOw the result as a dataframe, ALso, calculate the overall median and IQR

import pandas as pd

# Assuming 'df' is your DataFrame from the previous code
# Replace 'A1', 'A2', 'A3', 'A4', 'A5' with your actual column names if different.

def calculate_median_iqr(df, columns):
    results = []
    for col in columns:
        median = df[col].median()
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        results.append([col, median, iqr])

    results_df = pd.DataFrame(results, columns=['Column', 'Median', 'IQR'])

    overall_median = df[columns].median().median()
    overall_q1 = df[columns].quantile(0.25).quantile(0.25)
    overall_q3 = df[columns].quantile(0.75).quantile(0.75)
    overall_iqr = overall_q3 - overall_q1
    overall_results = pd.DataFrame({'Column': ['Overall'], 'Median': [overall_median], 'IQR':[overall_iqr]})

    return pd.concat([results_df, overall_results], ignore_index=True)


# Example usage
columns_to_analyze = ['A1', 'A2', 'A3', 'A4', 'A5']  # Replace with your column names
try:
    results_dataframe = calculate_median_iqr(df, columns_to_analyze)
    print(results_dataframe)
except KeyError as e:
    print(f"Error: Column '{e}' not found in the DataFrame. Please check the column names.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# prompt: calculate the net agreement rating. the formula is Percentage [(Strongly Agree + Agree) - (Strongly Disagree + Disagree)]. Please note that 4 is Strongly Agree, 3 for Agree, 2 for Disagree, and 1 for Strongly Disagree. The column name is A1, A2,A3,A4,A5,. Also, compute the overall net agreement rating. display the result in two decimal places only

def calculate_net_agreement(df, columns):
    results = []
    for col in columns:
        try:
            strongly_agree = df[df[col] == 4][col].count()
            agree = df[df[col] == 3][col].count()
            strongly_disagree = df[df[col] == 1][col].count()
            disagree = df[df[col] == 2][col].count()
            net_agreement = ((strongly_agree + agree) - (strongly_disagree + disagree)) / len(df) * 100
            results.append([col, "{:.2f}".format(net_agreement)])
        except KeyError:
            print(f"Column '{col}' not found in the DataFrame.")
            results.append([col, "N/A"])  # Or handle the error as needed
        except Exception as e:
            print(f"An error occurred while processing column '{col}': {e}")
            results.append([col, "Error"])

    overall_strongly_agree = sum(df[c] == 4 for c in columns).sum()
    overall_agree = sum(df[c] == 3 for c in columns).sum()
    overall_strongly_disagree = sum(df[c] == 1 for c in columns).sum()
    overall_disagree = sum(df[c] == 2 for c in columns).sum()
    overall_net_agreement = ((overall_strongly_agree + overall_agree) - (overall_strongly_disagree + overall_disagree)) / (len(columns) * len(df)) * 100
    results.append(["Overall", "{:.2f}".format(overall_net_agreement)])

    return pd.DataFrame(results, columns=["Column", "Net Agreement (%)"])

# Example usage
columns_to_analyze = ['A1', 'A2', 'A3', 'A4', 'A5']  # Replace with your actual column names
net_agreement_df = calculate_net_agreement(df, columns_to_analyze)
net_agreement_df
```
